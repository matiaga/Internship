# Assigment - 1 - Web Scrapping


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def del_specials_chars(text):
    return re.sub(r'^[\n\xa0]+|[\n\xa0]+$', '', text)

def get_header_tags_from(url): 
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Find all header tags
    header_tags = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6" ])

#     header_texts = [del_specials_chars(tag.text) for tag in header_tags]
    header_texts = [tag.text.strip() for tag in header_tags]
   
    return pd.DataFrame({'Headers': header_texts})

url = 'https://wikipedia.org'
get_header_tags_from(url)
```

```python
def get_list_presidents(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    
    president_listings  = soup.find_all('div', class_="presidentListing")
    
    # Retrieving the names and terms
    names = [div.find('h3').text.strip() for div in president_listings ]
    term_of_offices = [div.find('span', class_='terms').next_sibling.strip() for div in president_listings]
    
    # Checking length of lists
    if(len(names) == len(term_of_offices)):
        return pd.DataFrame({'Name': names, 'Term of office': term_of_offices})

    return pd.DataFrame()

url = "https://presidentofindia.nic.in/former-presidents.htm"
get_list_presidents(url)
```

```python
def get_10_from_container(cont): 
    ''' Return: A DataFrame  '''
    cat = [cont.find('h4').text for i in range(10)]

    f_pos = cont.find('span', class_="rankings-block__pos-number").text.strip()
    f_player = cont.find('div', class_="rankings-block__banner--name").text
    f_team = cont.find('span', class_="rankings-block__banner--nation").next_sibling
    f_rating = cont.find('div', class_="rankings-block__banner--rating").text
    
    table = cont.find('table', class_="table rankings-card-table")
    
    pos = [f_pos] + [ span.text.strip() for span in cont.findAll('span', class_="rankings-table__pos-number")]
    player = [f_player] + [ td.a.text for td in cont.findAll('td', class_="table-body__cell name")]
    team = [f_team] + [span.text for span in cont.findAll('span', class_="table-body__logo-text")]
    rating = [f_rating] + [td.text for td in cont.findAll('td', class_="table-body__cell u-text-right rating")]

    return pd.DataFrame({
            'Category': cat, 
            'Position': pos,
            'Player': player,
            'Team': team,
            'Rating': rating
        })

def get_rankings(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    
    containers = soup.findAll('div', class_="rankings-block__container")[0:3]
    
    df = pd.DataFrame()
    
    for cont in containers:
        df_i = get_10_from_container(cont)
        df = pd.concat([df, df_i])

    return df


url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi"
df = get_rankings(url)

all_rounders = df[df['Category'] == 'ODI All-Rounder Rankings']
print('*' * 70, "\n a) ODI All-Rounder Rankings:","\n", '*' * 69)
print(all_rounders)

batsmen = df[df['Category'] == 'ODI Batting Rankings']
print('*' * 70, "\n b) ODI Batting Rankings:","\n", '*' * 69)
print(batsmen)

bowlers = df[df['Category'] == 'ODI Bowling Rankings']
print('*' * 70,"\n c) ODI Bowling Rankings: ","\n", '*' * 69)
print(bowlers)
```

```python
def get_news(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    
    # Retriving the news
    news_divs = soup.find_all('div', class_=["RiverPlusCard-cardLeft","RiverPlusCard-breakerCardContainer","SecondaryCard-container"] )
    
    headline_divs = [new.find('div', class_=["RiverHeadline-headline RiverHeadline-hasThumbnail", "SecondaryCard-headline"]) for new in news_divs]
    times_divs = [new.find('span', class_="RiverByline-datePublished") for new in news_divs ]
    
    headlines = [div.text for div in headline_divs]
    links = [div.find_all('a')[1]['href']  if (len(div.find_all('a')) > 1) else div.find_all('a')[0]['href'] for div in headline_divs ]
    times = [ time.text if time!= None else None for time in times_divs]
    
#     print("News divs: ",len(news_divs))
#     print("Headline divs:",len(headline_divs))
#     print("times :", len(times))
    
    if (len(headlines) == len(links) and len(links) == len(times)):
        return pd.DataFrame({'Headline': headlines, 'Time': times, 'News Link': links})
    
    return pd.DataFrame()
    
    
url = "https://www.cnbc.com/world/?region=world"
df = get_news(url)
display(df)
```

```python
def get_articles_ai(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    articles = soup.find_all('article')

    title = [art.find('h2').text for art in articles]
    author = [ art.find('span', class_="sc-1w3fpd7-0 dnCnAO").text for art in articles]
    paper_url = [art.find('a')['href'] for art in articles]
    published_date = [art.find('span', class_="sc-1thf9ly-2 dvggWt").text.strip() for art in articles]

    if(len(title) == len(author) and len(paper_url) == len(published_date) and len(title)== len(paper_url)):
        df = pd.DataFrame({'Paper Title': title,'Authors': author, 'Published Date': published_date, 'Paper URL': paper_url})
    else:
        df = pd.DataFrame()
    return df

url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
get_articles_ai(url)
```

```python
def get_restaurants(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    rest_name = [tag.text.strip() for tag in soup.find_all('a', class_="restnt-name ellipsis")]
   
    cuisine_spans = soup.find_all('span', class_="double-line-ellipsis")
    cuisine = [ span.text.split('|')[1].strip() if len(span.text.split('|')) > 1 else '' for span in cuisine_spans]
   
    location = [div.text for div in soup.find_all('div', class_="restnt-loc ellipsis")]
    img_url = [img.get('data-src') for img in soup.find_all('img', class_="no-img")]
    rating = [div.text for div in soup.find_all('div', class_="restnt-rating rating-4")]
    
    return pd.DataFrame({'Restaurant name': rest_name ,
                         'Cuisine': cuisine, 
                         'Location': location, 
                         'Ratings': rating, 
                         'Image URL': img_url })

url = "https://www.dineout.co.in/delhi-restaurants/buffet-special"
get_restaurants(url)
```
